```python
import pandas as pd
from ydata_profiling import ProfileReport
import glob
```

###Profiling de TODOS os CSV do ENADE 2021/2017/2014

Geração de páginas .html de cada um desses arquivos. Os arquivos estão disponíveis para download no repositório da disciplina.

Para um primeiro momento, selecionei apenas o arq1 e arq2 para avaliação, pois eles possuiam dados que eram inputados pelos concluintes. Os demais arquivos não achei necessário em primeiro momento fazer uma análise mais aprofundada, porque eram de uma única escolha e era necessário escolher algo previamente estabelecido na prova.

```python
arquivos_enade = glob.glob('/content/drive/MyDrive/EDM 2025.1/DATASETS/ENADE/2014/*.txt')

for arquivo in arquivos_enade:
    print(f'Gerando profiling para: {arquivo}')

    df = pd.read_csv(arquivo, sep=';', encoding='latin1')

    nome_relatorio = arquivo.split('/')[-1].replace('.txt', '_profiling.html')

    profile = ProfileReport(df, title=f'Profiling - {arquivo}', explorative=True)
    profile.to_file(nome_relatorio)
```

### Profiling de TODOS os CSV do Censo Superior

```python
anos = ['2014', '2017', '2021']
base = '/content/drive/MyDrive/EDM 2025.1/DATASETS/CENSO'
```

```python
for ano in anos:
    arquivos = glob.glob(f'{base}/{ano}/*.CSV')
    print(f'Ano {ano} tem {len(arquivos)} arquivos encontrados')
```

Não é viável fazer o profilling do conjunto de dados todo. Muito pesado (145mb) e passou cerca de 15min rodando apenas 1 ano. Tem muitos cursos e muitos cursos duplicados e com inconsistência. O profiling do Censo vai ser feito após o devido tratamento para pegar apenas cursos do Nordeste, em um outro colab.

```python
for ano in anos:
    arquivos = glob.glob(f'{base}/{ano}/*.CSV')
    for arq in arquivos:
        try:
            df = pd.read_csv(arq, sep=';', encoding='latin1')
            nome = arq.split('/')[-1].replace('.csv', '_profiling.html')
            ProfileReport(df, title=nome, explorative=True).to_file(nome)
            print(f'✅ {nome}')
        except Exception as e:
            print('Erro')
```
